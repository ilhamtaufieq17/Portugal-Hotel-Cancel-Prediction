### **Business Context**  
In the hospitality industry, customer cancellations can have a profound impact on a hotel's revenue, operational efficiency, and overall business performance. When a customer cancels a booking, particularly at the last minute, it often leaves the hotel with limited time to rebook the room. This not only results in direct revenue losses but also disrupts the hotel's ability to manage its resources effectively. Rooms that remain unoccupied represent missed opportunities for income, and when cancellations happen frequently, the cumulative financial impact can be significant.

Moreover, the repercussions of frequent cancellations extend beyond just lost revenue. They can tarnish the hotel's reputation, especially if the cancellations create inconsistencies in service delivery or lead to overbooking as a countermeasure. Other guests might experience reduced satisfaction if the hotel's resources are stretched thin or if they face complications due to changes in room availability. This, in turn, can lead to negative reviews, reduced customer loyalty, and a weakened brand image.

### **False Predicition:**

In the context of developing a machine learning model to predict hotel booking cancellations, metric evaluation plays a crucial role in assessing the model's performance and its potential impact on hotel operations and revenue.

|                    | **Not Cancelled (0)**            | **Cancelled (1)**             |
|--------------------|-------------------------|-----------------------|
| **Not Cancelled (0)**   | **True Negative (TN)**  | **False Positive (FP)**   |
|                    | Prediction: No Churn    | Prediction: Churn     |
|                    | Actual: No Churn        | Actual: No Churn      |
| **Cancelled (1)**      | **False Negative (FN)**     | **True Positive (TP)**    |
|                    | Prediction: No Churn    | Prediction: Churn     |
|                    | Actual: Churn           | Actual: Churn         |


`**Type 1 error**: False Positive (FP)`: The model predicts that a booking will be canceled, but in reality, the booking is not canceled.

`Consequences`: 

In the case of a Type 1 error (False Positive) in predicting hotel booking cancellations, the hotel would incur unnecessary costs and lose potential revenue. For example, if the hotel prepares a room for rebooking based on an incorrect prediction, the **operational cost** might be **€42.50**.

Additionally, if the room could have been booked at an **average rate of €170.00 per night**, with a **70% probability of being occupied**, the potential revenue loss would be **€119.00**. If the false prediction leads to the loss of a customer, **the Customer Acquisition Cost (CAC) to replace them could be €85.00**. Moreover, if the dissatisfied customer never returns, the **Customer Lifetime Value (CLV) loss could be €425.00**.

Overall, the total loss per Type 1 error could be around **€671.50**.


The total loss cost for a Type 1 error (False Positive) can be calculated using the following formula:

$$
\text{Total Loss Cost} = \text{Operational Cost} + (\text{Average Room Rate (ARR)} \times \text{Occupancy Probability}) + \text{Customer Acquisition Cost (CAC)} + \text{Customer Lifetime Value (CLV) Loss}
$$

Substituting the values:

$$
\text{Total Loss Cost} = \text{€42.50} + (\text{€170} \times 0.70) + \text{€85} + \text{€425}
$$

$$
\text{Total Loss Cost} = \text{€42.50} + \text{€119} + \text{€85} + \text{€425} = \text{€671.50}
$$

`**Type 2 error**: False Negative (FN)`: The model predicts that a booking will not be canceled, but in reality, the booking is canceled.

`Consequences`: 

In the case of a Type 2 error (False Negative) in predicting hotel booking cancellations, the hotel would lose potential revenue and miss opportunities to mitigate the impact of cancellations. For example, if the room could have been booked at an **average rate of €170 per night**, with a **70% probability of being occupied**, the potential revenue loss would be **€119**.

Additionally, if the hotel had overbooked to cover expected cancellations, the compensation cost might add **€50**.

Overall, the total loss per Type 2 error could be around **€169**.

The total loss cost for a Type 2 error (False Negative) can be calculated using the following formula:

$$
\text{Total Loss Cost (Type 2 Error)} = \text{Lost Revenue} + \text{Overbooking Compensation}
$$

Substituting the values:

$$
\text{Total Loss Cost (Type 2 Error)} = (\text{ARR} \times \text{Occupancy Probability}) + \text{Overbooking Compensation}
$$

$$
\text{Total Loss Cost (Type 2 Error)} = (€170 \times 0.70) + €50 = €119 + €50 = €169
$$

**Project Limitation**

While this project aims to optimize the hotel's booking management by predicting which bookings are likely to be canceled, there are several limitations to consider:

1. **Data Availability**: The model is built on historical booking and cancellation data, which may not fully capture future booking behavior, especially if external factors change (e.g., economic conditions, travel trends, or customer preferences).

2. **Cost Assumptions**: The cost calculations for operational expenses, customer acquisition, and overbooking compensation are based on average estimates. These costs may vary depending on the hotel's specific arrangements, location, and scale of operations.

3. **Customer Segmentation**: The model assumes that all customers exhibit similar booking and cancellation patterns. However, there might be significant variations among different customer segments (e.g., business travelers vs. tourists) that the model may not capture accurately.

4. **Operational Strategies**: The model considers specific strategies, such as overbooking and last-minute promotions. If other strategies (e.g., dynamic pricing, loyalty programs) are used, the cost and effectiveness might differ, which is not accounted for in this analysis.

These limitations highlight areas where the model's predictions might be less accurate or where further refinement is needed to ensure broader applicability.

```python
class FeatureCreation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        # Ensure X is a DataFrame
        X = pd.DataFrame(X)
        X['total_service_requests'] = X['total_of_special_requests'] + X['required_car_parking_spaces']
        X['cancellation_history'] = np.where(X['previous_cancellations'] > 0, 1, 0)
        X['change_request_rate'] = X['booking_changes'] / (X['days_in_waiting_list'] + 1)
        X['special_request_intensity'] = X['total_of_special_requests'] / (X['total_of_special_requests'] + X['required_car_parking_spaces'] + 1)
        X['is_long_wait'] = np.where(X['days_in_waiting_list'] > 7, 1, 0)
        return X
```

```python
def loss_cost(y_pred, y_true):
    fp_cost = np.where((y_pred == 1) & (y_true == 0), 671.5, 0)
    fn_cost = np.where((y_pred == 0) & (y_true == 1), 169, 0)
    return np.sum([fp_cost, fn_cost])
```

The loss_cost function calculates a custom loss by assigning specific costs to false positives (671.5) and false negatives (169) based on model predictions compared to true labels. In the context of the confusion matrix, false positives (FP) occur when the model incorrectly predicts a positive outcome (1) for a negative instance (0), while false negatives (FN) occur when the model fails to predict a positive outcome for an actual positive instance.

**Focus on Precision**:
Precision is a metric that reflects how many of the positive predictions made by the model are actually correct. It is calculated as:

$$
\text{Precision} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Positives (FP)}}
$$
​	
 

Given this, the loss_cost function directly penalizes false positives (which reduce precision) with a higher cost of 671.5, emphasizing the importance of minimizing incorrect positive predictions. It also penalizes false negatives with a cost of 169, acknowledging the impact on recall but assigning a lower weight.

By summing the costs of false positives and false negatives, the function returns a total loss that reflects the overall impact of these misclassifications on the model’s performance. This loss function is particularly useful in scenarios where precision is a key focus, as it assigns a higher penalty to false positives, encouraging the model to make more accurate positive predictions.
